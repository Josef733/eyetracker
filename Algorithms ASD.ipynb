```python
#ASD Data

x_val = []
y_val = []
dur = []

duration_diff = []

point_disp = []  
x_point_len = []
y_point_len = []

point_velo_asd = []


for i in range(1,301):
    
    x_val3 = []
    y_val3 = []
    dur3 = []
    
    x_val2 = []
    y_val2 = []
    dur2 = []
    
    duration_diff2 = []
    point_disp2 = []
    
    x_point_len2 = []
    y_point_len2 = []
    
    participant = i
    test = pd.read_csv("data\ASD\ASD_scanpath_"+ str(participant) + ".txt",sep=",")
    test.columns = ["Idx", "x", "y", "duration"]

    x_val3 = test.iloc[:, 1:2].values
    y_val3 = test.iloc[:, 2:3].values
    dur3 = test.iloc[:, 3:4].values  

    x_val2 = np.array(x_val3).sum(axis=1)
    y_val2 = np.array(y_val3).sum(axis=1)
    dur2 = np.array(dur3).sum(axis=1)
    
    
    for i in range(len(x_val2)):
        if math.isnan(x_val2[i]):
            pass
        elif math.isnan(y_val2[i]):
            pass
        else:
            x_val.append(x_val2[i])
            y_val.append(y_val2[i])
            dur.append(dur2[i])
    
    
    # Feature Duration
    
    
    for i in range(len(x_val2)-1):
    
        if math.isnan(dur2[i+1] - dur2[i]):
            pass
        else:
            duration_diff2.append(dur2[i]/1000)
        
        
    # Feature Displacement    
       
        
    for i in range(len(x_val2)-1):
    
        x_point_len2.append(x_val2[i+1] - x_val2[i])
    
        y_point_len2.append(y_val2[i+1] - y_val2[i])
        
        point_disp2.append(np.sqrt(x_point_len2[i]**2 + y_point_len2[i]**2))

    
    #Velocity
    
    
    for i in range(len(point_disp2)):
        
        point_velo_asd.append(point_disp2[i]/(duration_diff2[i]))


print(len(x_val))
print(len(y_val))
print(len(dur))

print(len(duration_diff))
print(len(point_disp))
print(len(point_velo_asd))
#print(point_velo)    
        
#sb.histplot(point_velo, element = 'poly', fill= False)
plt.hist(point_velo_asd,bins = 200) 
plt.xlim(xmin=0, xmax = 50000)
plt.title("Velocity (All Participants) (ASD)")
plt.xlabel("Velocity (Pixels/Second)")
plt.ylabel("Frequency")
plt.show()

mean_asd = statistics.mean(point_velo_asd)
print("Mean Velocity (ASD): " + str(mean_asd))
```

```python
#K-Ratio

def k_ratio(dat_type):
    
    import time

    start = time.time()

    k_ratios = []
    v_thres = []

    for i in range(4000):
    
        sac_timestamp = []

        classifier = []

        velocity_threshold = i + 10
    
        #i-vt
    
        for k in range (len(dat_type)):
            if dat_type[k] >= 0 and dat_type[k] < velocity_threshold:
                classifier.append("fixation")
        
            elif dat_type[k] >= velocity_threshold:
                sac_timestamp.append("0")
                classifier.append("saccade")
        
            else:
                classifier.append("outlier")
    
        #print(len(classifier))
        #print(len(sac_timestamp))
    
        #k-ratio
        # P is the probability of being a saccade state

        P = len(sac_timestamp)/len(classifier)

        #print(P)

        # N*(P)*(1-P)

        p_ind = len(classifier)*(P)*(1-P)

        # p_ind = (P)*(1-P)          (paper formula)

        #print(p_ind)

        #number of transitions

        p_emp = 0

        for j in range (len(classifier)-1):
            if (classifier[j] != classifier[j+1]):
                p_emp = p_emp + 1
            else:
                pass

        #print(p_emp) 

        #k-ratio

        k_ratio = p_emp / p_ind
    
        #print(k_ratio)   
    
        k_ratios.append(k_ratio)
        v_thres.append(i + 1)

    plt.plot(v_thres,k_ratios, c="magenta",label="K-Ratio")
    plt.legend()
    plt.ylabel("K-Ratio Value")
    #plt.ylim(-250, 1250)
    #plt.xlim(-200, 2000)
    plt.xlabel("Velocity Threshold")
    plt.title("K-Ratios for I-VT")
    plt.savefig("k_rat_ivt")
    plt.show()


    end = time.time()
    total_time = end - start
    print("\n"+ str(total_time))

    low = 2.0
    thres = 0
    for m in range(len(k_ratios)):
       if low > k_ratios[m]:
        low = k_ratios[m]
        thres = v_thres[m]
    
    print("K-Ratio: " + str(low))
    print("Threshold: " + str(thres))
```
